```python
def compare_by_sum(point):
    return point[0] + point[1]
```

```python
points = [(1, -1), (2, 3), (-10, 15), (10, 9), (7, 18), (1, 5), (2, -4)]
```

```python
df.head()
```

### 1.1. Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
# df['Cpu_Company'] = df['Cpu']
df['Cpu'].value_counts()
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split()[0])
df['Cpu_Company'].value_counts()
```

### 1.2. Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
df['Memory'].value_counts()
```

```python
def gb(s):
    s = s.split()[0]
    if s.endswith('TB'):
        x = int(float(s.replace('TB', ''))) * 1024
    elif s.endswith('GB'):
        x = int(s.replace('GB', ''))
    return x

df['Memory_Amount'] = df['Memory'].apply(gb)
df['Memory_Amount'].value_counts()
```

### 1.3. Создать новый признак Memory_Type, который будет содержать тип носителя (HDD/SDD/др.)

```python
df['Memory_Type'] = df['Memory'].apply(lambda x: x.split()[-1])
df['Memory_Type'].value_counts()
```

### 1.4. Удалите признаки Memory и ScreenResolution

```python
# df.drop(columns = ['Memory', 'ScreenResolution'], inplace = True)
# df.drop(['Memory', 'ScreenResolution'], axis = 1, inplace = True)
df1 = df.drop(['Memory', 'ScreenResolution'], axis = 1)
df1
```

### 4.3. Постройте таблицу с подсчётом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ

```python
df.pivot_table(
    index = 'Cpu_Company',
    columns = 'Ram',
    aggfunc = 'count',
    values = 'laptop_ID',
    fill_value = '-'
)
```

### 4.4. Постройте таблицу с подсчётом средней стоимости ноутбуков в данных в зависимости от опрационной системы и GB памяти

#### С помощью pivot_table

```python
df.pivot_table(
    index = 'OpSys',
    columns = 'Ram',
    aggfunc = 'mean',
    values = 'Price_euros',
    fill_value = '-'
)
```

#### С помощью crosstab

```python
pd.crosstab(
    index = df['OpSys'],
    columns = df['Ram'],
    aggfunc = 'mean',
    values = df['Price_euros']
).fillna('-')
```

# 3-й способ: Подсвечивание данных

```python
data = pd.crosstab(
    index = df['Memory_Type'],
    columns = df['Memory_Amount']
)
```

```python
def gold_max(x):
    m = x.to_numpy().max() # в numpy, т.к. мы хотим макс от всех значений
    color = {
        True: "background-color: #c78f2e",
        False: ""
    }
    is_max = (x == m).replace(color)
    return is_max
```

```python
data.style.apply(gold_max, axis = None) # Максимум из всех данных: axis = None
```

```python
data.to_numpy()
```

```python
data.to_numpy().max()
```

```python
def highlight_max(x):
    return np.where(x == np.max(x.to_numpy()), f"color: red ;", "")
```

```python
data.style.apply(highlight_max, axis = 0) # Максимум по столбцам: axis = 0
```

```python
def blue_max(x):
    m = x.to_numpy().max() # в numpy, т.к. мы хотим макс от всех значений
    color = {
        True: f"color: blue ;",
        False: ""
    }
    is_max = (x == m).replace(color)
    return is_max
```

```python
data.style.apply(blue_max, axis = 1) # Максимум по строкам: axis = 1
```
