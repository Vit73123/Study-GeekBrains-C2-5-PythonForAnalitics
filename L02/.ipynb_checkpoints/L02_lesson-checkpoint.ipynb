## Series

Индексы могут проставляться вручную

```python
a = pd.Series([4, 7, 6, 3, 9],
    index = ['one', 'two', 'three', 'four', 'five'])
a
```

Индексы могут проставляться автоматически

```python
a = pd.Series([4, 7, 6, 3, 9])
a
```

Индексов должно быть столько же, сколько и элементов

```python
a = pd.Series([4, 7, 6, 3, 9, 3],
    index = ['one', 'two', 'three', 'four', 'five'])
a
```

Значения статистики:
* Count - количество непропущенных объектов (там, где нет _nan-значений_)
* mean - среднее арифметическое по столбцу
* std - стандартное отклонение по столбцу
* min - минимальное значение в столбце  

Квантили:
* 25%
* 50% - он же медиана
* 75%  

* max - максимальное значение в столбце

## Фильтрация
Основана на булевых масках

```python
df['Gender'] == 'Male' # Булева маска - "Если Maile, то True, иначе False"
```

```python
male = df[df['Gender'] == 'Male'] # Задать фильтр по Булевой маске
male                              # Вывести фильтр (отфильтрованные данные DataFrame)
```

```python
(df['Gender'] == 'Female') & (df['NumOfProducts'] >= 3)
```

### Логическое И

```python
df[(df['Gender'] == 'Female') & (df['NumOfProducts'] >= 3)]
```

## Логическое ИЛИ

```python
(df['HasCrCard'] == 'Female') | (df['NumOfProducts'] >= 3)
```

## Логическое НЕ

```python
~(df['Geography'] == 'Spane')
```

```python
df[~(df['Geography'] == 'Spane')]
```

### Принадлежность множеству - isin

```python
df[df['Geography'].isin(['France', 'Germany'])]
```
